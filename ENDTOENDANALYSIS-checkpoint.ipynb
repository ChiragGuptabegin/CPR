```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import tensorflow as tf

```

```python
df= pd.read_csv('test1.csv')
df = df[0::6]
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date']= pd.to_datetime(df['date'], infer_datetime_format=True)
```

```python
df.set_index('date')
```

```python
df_input=df[['CPRAVG','PROP','BUT','ETHYLENE','AI12201M','FLW','COT','TEMP','PRESS','O2','DRAFT','ETH']]
```

```python
df_input
```

```python
scaler= MinMaxScaler(feature_range=(0,1))
data_scaled= scaler.fit_transform(df_input)
```

```python
data_scaled
```

```python
features=data_scaled
target=data_scaled[:,0]
```

```python
TimeseriesGenerator(features,target,length=1,sampling_rate=1,batch_size=1)[0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=123,shuffle=False)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
win_length=80
batch_size=256
num_features=12
train_generator=TimeseriesGenerator(x_train,y_train,length=win_length,sampling_rate=1,batch_size=batch_size)
test_generator=TimeseriesGenerator(x_test,y_test,length=win_length,sampling_rate=1,batch_size=batch_size)

```

```python
train_generator[0]
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
import tensorflow as tf


model= tf.keras.Sequential()
model.add(LSTM(30, input_shape=(win_length, num_features),return_sequences=True))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(LSTM(15,return_sequences=False))   
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(1))
 
```

```python
model.summary()
```

```python
early_stopping= tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2,mode='min')
model.compile(loss=tf.losses.MeanSquaredError(),optimizer=tf.optimizers.Adam(),metrics=[tf.metrics.MeanAbsoluteError()])
history=model.fit(train_generator,epochs=50,validation_data=test_generator,shuffle=False,callbacks=[early_stopping])
```

```python
model.evaluate(test_generator,verbose=0)
```

```python
predictions=model.predict(test_generator)
```
